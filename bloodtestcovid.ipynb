A lot of complex relationships present. Let's plot the following:

* AST (aspartame transferase) vs ALT (alanine transaminase)
* RBC (red blood count) vs HGB (hemoglobin)
* NE (neutrophils) vs LY (lymphocytes)

All three have high correlation coefficients. Looking at the bottom right of the correlation matrix, a strong relationship exists between NE,	LY,	MO,	EO.

```python
#Aspartame Transferase vs ALanine Transaminase concentrations
plt.title("Aspartame Transferase vs Alanine Transaminase Concentrations")
plt.xlabel('Aspartame Transferase')
plt.ylabel('Alanine Transaminase')
plt.scatter(bti['AST'].as_data_frame(), bti['ALT'].as_data_frame())
```

Lower amounts of aspertame transferase --> lower levels of alanine transaminase, more scatter/spread as graph goes away from origin.

```python
plt.title("Red Blood Count vs Hemoglobin Concentrations")
plt.xlabel('Red Blood Count')
plt.ylabel('Hemoglobin')
plt.scatter(bti['RBC'].as_data_frame(), bti['HGB'].as_data_frame())
plt.xlim([0,10])
plt.ylim([0,30])
plt.show()
```

Clustering, with a positive correlation

```python
plt.title("Neutrophil vs Lymphocyte Concentrations")
plt.xlabel('Neutrophil')
plt.ylabel('Lymphocytes')
plt.scatter(bti['NE'].as_data_frame(), bti['LY'].as_data_frame())
```

Next step -- use AutoML to find the best ML algorithm that can classify whether a certain set of blood test results is indicative of COVID-19

```python
trainbti, testbti = bti.split_frame(ratios=[0.8], seed = 1)
trainbti['target'] = trainbti['target'].asfactor()
trainbti
```

```python
t = H2OAutoML(max_models=15, balance_classes=True, seed = 1)
```

```python
t.train(training_frame=trainbti, validation_frame=testbti, y='target')
```

```python
lb = t.leaderboard
lb
```

```python
bestModel = t.get_best_model()
bestModel
```

```python
predictions = bestModel.predict(testbti)
predictions
```

```python
bestModelPath = h2o.save_model(model=bestModel, path='/Users/devam/PycharmProjects/AutoMLEDA/internships/bestModel', force=True)
```
